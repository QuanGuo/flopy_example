### Create the Flopy Model Objects

We are creating a square model with a specified head equal to `h1` along all boundaries. The head at the cell in the center in the top layer is fixed to `h2`. First, set the name of the model and the parameters of the model: the number of layers `Nlay`, the number of rows and columns `N`, lengths of the sides of the model `L`, aquifer thickness `H`, hydraulic conductivity `k`

```python
name = "mf6lake"
h1 = 0    
# h2 = -90
Nlay = 10
N = 128
L = 320.0
H = 10.0
```

One big difference between MODFLOW 6 and previous MODFLOW versions is that MODFLOW 6 is based on the concept of a simulation.  A simulation consists of the following:

* Temporal discretization (TDIS)
* One or more models (GWF is the only model supported at present)
* Zero or more exchanges (instructions for how models are coupled)
* Solutions

For this simple lake example, the simulation consists of the temporal discretization (TDIS) package (TDIS), a groundwater flow (GWF) model, and an iterative model solution (IMS), which controls how the GWF model is solved.

```python
# Create the Flopy simulation object
sim = flopy.mf6.MFSimulation(
    sim_name=name, exe_name="mf6", version="mf6", sim_ws=workspace
)

# Create the Flopy temporal discretization object
tdis = flopy.mf6.modflow.mftdis.ModflowTdis(
    sim, pname="tdis", time_units="DAYS", nper=1, perioddata=[(1.0, 1, 1.0)]
)

# Create the Flopy groundwater flow (gwf) model object
model_nam_file = f"{name}.nam"
gwf = flopy.mf6.ModflowGwf(sim, modelname=name, model_nam_file=model_nam_file)

# Create the Flopy iterative model solver (ims) Package object
ims = flopy.mf6.modflow.mfims.ModflowIms(sim, pname="ims", complexity="SIMPLE")
```

Now that the overall simulation is set up, we can focus on building the groundwater flow model.  The groundwater flow model will be built by adding packages to it that describe the model characteristics.

Define the discretization of the model. All layers are given equal thickness. The `bot` array is build from `H` and the `Nlay` values to indicate top and bottom of each layer, and `delrow` and `delcol` are computed from model size `L` and number of cells `N`. Once these are all computed, the Discretization file is built.

```python
# Create the discretization package
bot = np.linspace(-H / Nlay, -H, Nlay)
delrow = delcol = L / (N - 1)
dis = flopy.mf6.modflow.mfgwfdis.ModflowGwfdis(
    gwf,
    pname="dis",
    nlay=Nlay,
    nrow=N,
    ncol=N,
    delr=delrow,
    delc=delcol,
    top=0.0,
    botm=bot,
)
```

```python
# Create the initial conditions package
start = h1 * np.zeros((Nlay, N, N))
ic = flopy.mf6.modflow.mfgwfic.ModflowGwfic(gwf, pname="ic", strt=start)
```

### create chd

```python
# Create the constant head package.
# List information is created a bit differently for
# MODFLOW 6 than for other MODFLOW versions.  The
# cellid (layer, row, column, for a regular grid)
# must be entered as a tuple as the first entry.
# Remember that these must be zero-based indices!
chd_rec = []
# chd_rec.append(((0, int(N / 2), int(N / 2)), h2))
for layer in range(0, Nlay):
    for row_col in range(0, N):
        chd_rec.append(((layer, row_col, 0), h1))
        chd_rec.append(((layer, row_col, N - 1), h1))
        # if row_col != 0 and row_col != N - 1:
        #     chd_rec.append(((layer, 0, row_col), h1))
        #     chd_rec.append(((layer, N - 1, row_col), h1))
chd = flopy.mf6.modflow.mfgwfchd.ModflowGwfchd(
    gwf,
    pname="chd",
    maxbound=len(chd_rec),
    stress_period_data=chd_rec,
    save_flows=True,
)
```

```python
# The chd package stored the constant heads in a structured
# array, also called a recarray.  We can get a pointer to the
# recarray for the first stress period (iper = 0) as follows.
iper = 0
ra = chd.stress_period_data.get_data(key=iper)
ra
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_grf(psd_function, size, grid_step=1.0):
    # Frequency grid
    freq_x = np.fft.fftfreq(size, d=grid_step)
    freq_y = np.fft.fftfreq(size, d=grid_step)
    freq_grid_x, freq_grid_y = np.meshgrid(freq_x, freq_y)
    freq_radius_sq = freq_grid_x**2 + freq_grid_y**2

    # Calculate power spectral density
    psd = psd_function(freq_radius_sq)

    # Generate random field in frequency domain
    random_field_freq_domain = np.random.normal(size=(size, size)) + 1j * np.random.normal(size=(size, size))
    random_field_freq_domain *= np.sqrt(psd) / (size ** 2)

    # Inverse FFT to transform to spatial domain
    random_field = np.fft.ifft2(random_field_freq_domain)
    return np.real(random_field)

# PSD for Squared Exponential Kernel
def squared_exponential_psd(freq_radius_sq, length_scale=1.0, sigma=1.0):
    return sigma**2 * (2 * np.pi * length_scale**2) * np.exp(-2 * np.pi**2 * length_scale**2 * freq_radius_sq)

# PSD for Exponential Kernel
def exponential_psd(freq_radius_sq, length_scale=1.0, sigma=1.0):
    return 2 * sigma**2 * length_scale**2 / (1 + 4 * np.pi**2 * length_scale**2 * freq_radius_sq)

# Example: Generate GRF using Exponential PSD
size = 128
length_scale = 32
sigma = 1.0

# Choose the kernel here
kernel = "Exponential"  # Options: "SE", "Matern", "Exponential"

if kernel == "SE":
    psd_function = lambda r_sq: squared_exponential_psd(r_sq, length_scale, sigma)
elif kernel == "Exponential":
    psd_function = lambda r_sq: exponential_psd(r_sq, length_scale, sigma)
else:
    raise ValueError("Unknown kernel")

field = generate_grf(psd_function, size)

# Plot
plt.imshow(field, origin='lower')
plt.colorbar(label='Field Value')
plt.title(f'Stationary Gaussian Random Field - {kernel} Kernel')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process.kernels import Matern, RBF
def nystrom_method(X, kernel_type='RBF', n_components=10, length_scale=1.0, sigma=1.0, random_state=None):
    """
    Approximate eigenvalues and eigenvectors of the specified kernel matrix using the Nystrm method.

    Parameters:
    X: ndarray of shape (n_samples, n_features)
        The data points.
    kernel_type: str
        Type of kernel to use ('RBF' or 'Exponential').
    n_components: int
        Number of landmarks to use.
    length_scale: float
        The length scale of the kernel.
    sigma: float
        The variance of the kernel.
    random_state: int or None
        Random state for reproducibility.
    
    Returns:
    eigenvalues: ndarray
        Approximated eigenvalues of the kernel matrix, sorted in descending order.
    eigenvectors: ndarray
        Approximated eigenvectors of the kernel matrix.
    """
    rng = np.random.default_rng(random_state)
    n_samples = X.shape[0]

    indices = rng.choice(n_samples, size=n_components, replace=False)
    landmarks = X[indices]

    if kernel_type == 'RBF':
        kernel = RBF(length_scale=length_scale)
    elif kernel_type == 'Exponential':
        kernel = Matern(length_scale=length_scale, nu=0.5)
    else:
        raise ValueError("Unknown kernel type")

    W = kernel(landmarks)
    C = kernel(X, landmarks)
    M = np.dot(C.T, C)

    # Eigenvalue decomposition
    eigenvalues, eigenvectors_small = np.linalg.eigh(M)

    # Scale the eigenvalues
    eigenvalues *= n_samples / n_components

    # Approximate the eigenvectors for the large kernel matrix
    eigenvectors = np.dot(C, eigenvectors_small) / np.sqrt(eigenvalues)

    # Sort the eigenvalues and corresponding eigenvectors in descending order
    idx = eigenvalues.argsort()[::-1]
    eigenvalues_sorted = eigenvalues[idx]
    eigenvectors_sorted = eigenvectors[:, idx]

    return eigenvalues_sorted, eigenvectors_sorted

# Example usage
X = np.random.rand(128*128, 2)  # 1000 samples, 2 features

# For Gaussian kernel
eigenvalues_gaussian, eigenvectors_gaussian = nystrom_method(X, kernel_type='RBF', n_components=50, length_scale=0.2, sigma=10, random_state=42)

# For Exponential kernel
eigenvalues_exponential, eigenvectors_exponential = nystrom_method(X, kernel_type='Exponential', n_components=50, length_scale=0.2, sigma=1.0, random_state=42)

# Now you have both eigenvalues and eigenvectors for each kernel type
print(eigenvectors_gaussian.shape)
print(128*128)
```
